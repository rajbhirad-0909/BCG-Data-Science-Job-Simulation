# BCG X â€“ Task 2: Exploratory Data Analysis
This notebook contains the required EDA for the PowerCo churn project.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

client_path = 'client_data.csv'
price_path = 'price_data.csv'

client_df = pd.read_csv(client_path)
price_df = pd.read_csv(price_path)

client_df.head(), price_df.head()
```

## Data Types

```
client_df.info()
price_df.info()
```

## Missing Values

```
client_df.isnull().sum(), price_df.isnull().sum()
```

## Numeric Distributions

```
num_cols = client_df.select_dtypes(include=['int64','float64']).columns
client_df[num_cols].hist(figsize=(15,15))
plt.tight_layout()
plt.show()
```

## Churn Distribution

```
client_df['churn'].value_counts().plot(kind='bar')
plt.title('Churn Distribution')
plt.show()
```

## Merge Client & Price Data

```
latest_price = price_df.sort_values('price_date').groupby('id').tail(1)
merged_df = client_df.merge(latest_price, on='id', how='left')
merged_df.head()
```
